Right Triangle Pattern

```python
rows = 5

for i in range(1, rows + 1):
    print("*" * i)
```

Pyramid Pattern

```python
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("*" * (2 * i - 1))
```

Inverted Pyramid Pattern

```python
rows = 5

for i in range(rows, 0, -1):
    print(" " * (rows - i), end="")
    print("*" * (2 * i - 1))
```

Number Right Triangle

```python
rows = 5

for i in range(1, rows + 1):
    for j in range(1, i + 1):
        print(j, end="")
    print()
```

Number Pyramid Triangle

```python
rows = 5

for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    for j in range(1, i + 1):
        print(j, end="")
    for j in range(i - 1, 0, -1):
        print(j, end="")
    print()
```

Diamond Pattern

```python
rows = 5

# Upper part of the diamond
for i in range(1, rows + 1):
    print(" " * (rows - i), end="")
    print("*" * (2 * i - 1))

# Lower part of the diamond
for i in range(rows - 1, 0, -1):
    print(" " * (rows - i), end="")
    print("*" * (2 * i - 1))
```
